Model

```python
from sklearn.model_selection import train_test_split
x=df.drop('Cover_Type',axis=1)
y=df['Cover_Type']
x_train,x_test,y_train,y_test=train_test_split(x, y,test_size=0.3,random_state=42)
```

```python
from imblearn.under_sampling import RandomUnderSampler
random=RandomUnderSampler()
x_train,y_train=random.fit_resample(x_train,y_train)
```

```python
print(y_train.value_counts())
```

MODEL SELECTION

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import AdaBoostClassifier,GradientBoostingClassifier,RandomForestClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix,accuracy_score,f1_score,precision_score,classification_report
```

```python
scaler=StandardScaler()
x_trainscaled=scaler.fit_transform(x_train)
x_testscaled=scaler.fit_transform(x_test)
```

```python
log=LogisticRegression()
log.fit(x_trainscaled,y_train)
y_predl=log.predict(x_testscaled)
```

EVALUATION METRICS

```python
print("Accuracy:",accuracy_score(y_test,y_predl))
print("Precision:",confusion_matrix(y_test,y_predl))
print("ClassFication:",classification_report(y_test,y_predl))
sns.heatmap(confusion_matrix(y_test,y_predl),annot=True, fmt='d')
```

SVM

```python
svc=SVC()
svc.fit(x_trainscaled,y_train)
y_predl=svc.predict(x_testscaled)
print("Accuracy:",accuracy_score(y_test,y_predl))
print("Precision:\n",confusion_matrix(y_test,y_predl))
print("ClassFication:",classification_report(y_test,y_predl))
sns.heatmap(confusion_matrix(y_test,y_predl),annot=True, fmt='d')
```

```python
knn=KNeighborsClassifier()
knn.fit(x_trainscaled,y_train)
y_predl=knn.predict(x_testscaled)
print("Accuracy:",accuracy_score(y_test,y_predl))
print("Precision:",confusion_matrix(y_test,y_predl))
print("ClassFication:",classification_report(y_test,y_predl))
sns.heatmap(confusion_matrix(y_test,y_predl),annot=True, fmt='d')
```

```python
treemodel=DecisionTreeClassifier()
treemodel.fit(x_train,y_train)
y_predl=treemodel.predict(x_test)
print("Accuracy:",accuracy_score(y_test,y_predl))
print("Precision:",confusion_matrix(y_test,y_predl))
print("ClassFication:",classification_report(y_test,y_predl))
sns.heatmap(confusion_matrix(y_test,y_predl),annot=True, fmt='d')
```

```python
forestmodel=RandomForestClassifier()
forestmodel.fit(x_train,y_train)
y_predl=forestmodel.predict(x_test)
print("Accuracy:",accuracy_score(y_test,y_predl))
print("Precision:",confusion_matrix(y_test,y_predl))
print("ClassFication:",classification_report(y_test,y_predl))
sns.heatmap(confusion_matrix(y_test,y_predl),annot=True, fmt='d')
```

```python
adaboostmodel=AdaBoostClassifier()
adaboostmodel.fit(x_train,y_train)
y_predl=adaboostmodel.predict(x_test)
print("Accuracy:",accuracy_score(y_test,y_predl))
print("Precision:",confusion_matrix(y_test,y_predl))
print("ClassFication:",classification_report(y_test,y_predl))
sns.heatmap(confusion_matrix(y_test,y_predl),annot=True, fmt='d')
```

```python
Graboostmodel=GradientBoostingClassifier()
Graboostmodel.fit(x_train,y_train)
y_predl=Graboostmodel.predict(x_test)
print("Accuracy:",accuracy_score(y_test,y_predl))
print("Precision:",confusion_matrix(y_test,y_predl))
print("ClassFication:",classification_report(y_test,y_predl))
sns.heatmap(confusion_matrix(y_test,y_predl),annot=True, fmt='d')
```
